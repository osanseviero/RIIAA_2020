## Alineacion de posicion de objetos

Estimar precisamente la posicion de objetos es fundamental en varias industrias. Por ejemplo, para realidad aumentada o virtual, permite a los usuarios modificar el estado de algunas variables interactuando con objetos (por ejemplo, el volumen de un vaso en el escritorio del usuario).

![](https://storage.googleapis.com/tensorflow-graphics/notebooks/6dof_pose/task.jpg)

En este ejercicio, ilustraremos cómo estimar la rotación y traslación de objetos 3D. Veremos cómo hacer esto de dos maneras diferentes:


*   Optimización matemática.
*   Machine Learning. Entrenaremos una red neuronal sencilla para estimar precisamente la rotación y traslación de un objeto respecto a una pose de referencia.





````
from tensorflow import keras
from tensorflow.keras import layers
import matplotlib.pyplot as plt

from tensorflow_graphics.geometry.transformation import quaternion
from tensorflow_graphics.math import vector
from tensorflow_graphics.notebooks import threejs_visualization
from tensorflow_graphics.notebooks.resources import tfg_simplified_logo
import time

vertices = tfg_simplified_logo.mesh['vertices'].astype(np.float32)
faces = tfg_simplified_logo.mesh['faces']
num_vertices = vertices.shape[0]

threejs_visualization.triangular_mesh_renderer(tfg_simplified_logo.mesh, width=400, height=400)
````

### Red Neuronal

#### Definición de modelo.
Dado la posición 3D de todos los vértices de un mesh, nos gustaría tener un modelo que sea capaz de predecir la rotación parametrizada por un [cuaternión](https://www.tensorflow.org/graphics/api_docs/python/tfg/geometry/transformation/quaternion) (vector de 4 dimensiones) y traslación (vector de 3 dimensiones) del mesh respecto a una pose de referencia. 

Crearemos una red densa simple de 3 capas. Este modelo es muy simple y definitivamente no es óptimo, pero encontrar el modelo óptimo queda fuera del alcance de este ejercicio.

````
# Constructs the model.
model = keras.Sequential()
model.add(layers.Flatten(input_shape=(num_vertices, 3)))
model.add(layers.Dense(64, activation=tf.nn.tanh))
model.add(layers.Dense(64, activation=tf.nn.relu))
model.add(layers.Dense(7))
````

#### Función de pérdida

````
def pose_estimation_loss(y_true, y_pred):
  """Pérdida de estimación de posición usada en entrenamiento.

  Esta pérdida mide el promedio de la distancia^2 entre los vértices de un mesh
  posición destino y el mesh transformado al cuál se le aplica la predicción.

  Args:
    y_true: Valor real.
    y_pred: La predicción a evaluar.

  Returns:
    Un escalar conteniendo el valor de pérdida descrito arriba.
  """
  # y_true.shape : (batch, 7) -> (batch, 4) (batch, 3)
  y_true_q, y_true_t = tf.split(y_true, (4, 3), axis=-1)
  y_pred_q, y_pred_t = tf.split(y_pred, (4, 3), axis=-1)
  
  # [num_vertices, 3] -> [num_vertices, 1, 3]
  corners = tf.expand_dims(vertices, axis=1)

  # transformed_corners.shape: (num_vertices, batch, 3)
  transformed_corners = quaternion.rotate(corners, y_pred_q) + y_pred_t

  # recovered_corners.shape: (num_vertices, batch, 3)
  recovered_corners = quaternion.rotate(transformed_corners - y_true_t,
                                        quaternion.inverse(y_true_q))

  # vertex_error.shape: (num_vertices, batch)
  vertex_error = tf.reduce_sum((recovered_corners - corners)**2, axis=-1)

  return tf.reduce_mean(vertex_error)
````

````
optimizer = keras.optimizers.Adam()
model.compile(loss=pose_estimation_loss, optimizer=optimizer)
model.summary()
````

### Generación de datos

Ahora que tenemos el modelo definido, necesitamos datos para entrenarlo. Para cada ejemplo en el training set, se muestrean rotaciones y traslaciones 3D aleatorias y se aplican al objeto. Cada muestra consiste de todos los vértices transformados y la rotación y traslación inversa que lograrían revertir la rotación y traslación aplicada a la muestra.

````
def generate_training_data(num_samples):
  # random_angles.shape: (num_samples, 3)
  random_angles = np.random.uniform(-np.pi, np.pi,
                                    (num_samples, 3)).astype(np.float32)

  # Transforma notación de ángulo de Euler a cuaternión
  # random_quaternion.shape: (num_samples, 4)
  random_quaternion = quaternion.from_euler(random_angles)

  # random_translation.shape: (num_samples, 3)
  random_translation = np.random.uniform(-2.0, 2.0,
                                         (num_samples, 3)).astype(np.float32)

  # data.shape : (num_samples, num_vertices, 3)
  data = quaternion.rotate(vertices[tf.newaxis, :, :],
                           random_quaternion[:, tf.newaxis, :]
                          ) + random_translation[:, tf.newaxis, :]

  # target.shape : (num_samples, 4+3)
  target = tf.concat((random_quaternion, random_translation), axis=-1)

  return np.array(data), np.array(target)
````

````
num_samples = 10000

data, target = generate_training_data(num_samples)

print(data.shape)   # (num_samples, num_vertices, 3): los vértices del objeto
print(target.shape)  # (num_samples, 4+3): el cuaternión y la traslación
````

### Entrenamiento
¡Está todo listo para entrenar el modelo!

````
class ProgressTracker(keras.callbacks.Callback):
  def __init__(self, num_epochs, step=5):
    self.num_epochs = num_epochs
    self.current_epoch = 0.
    self.step = step
    self.last_percentage_report = 0

  def on_epoch_end(self, batch, logs={}):
    self.current_epoch += 1.
    training_percentage = int(self.current_epoch * 100.0 / self.num_epochs)
    if training_percentage - self.last_percentage_report >= self.step:
      print('Training ' + str(
          training_percentage) + '% complete. Training loss: ' + str(
              logs.get('loss')) + ' | Validation loss: ' + str(
                  logs.get('val_loss')))
      self.last_percentage_report = training_percentage
````

````
epochs = 100
pt = ProgressTracker(epochs)

history = model.fit(
    data,
    target,
    epochs=epochs,
    validation_split=0.2,
    verbose=0,
    batch_size=32,
    callbacks=[pt])
````

````
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.ylim([0, 1])
plt.legend(['loss', 'val loss'], loc='upper left')
plt.xlabel('Train epoch')
_ = plt.ylabel('Error [mean square distance]')
````

### Pruebas

¡La red está entrenada y lista para ser utilizada! Los resultados consisten de dos imágenes.


1.   La primera imagen consiste del objeto en la posición de descanso (color pastel limón) y el objeto rotado y trasladado. Esto nos permite observar efectivamente qué tan **diferentes** son las dos configuraciones.
2.   La segunda imagen también muestra el objeto en posición de descanso, pero, esta vez, se aplica la transformación predicha se aplica en el objeto rotado y trasladado. La expectativa es que ambos tengan una posición **similar**.

````
def transform_points(target_points, quaternion_variable, translation_variable):
  return quaternion.rotate(target_points,
                           quaternion_variable) + translation_variable
````

````
class Viewer(object):
  def __init__(self, my_vertices):
    my_vertices = np.asarray(my_vertices)
    context = threejs_visualization.build_context()
    light1 = context.THREE.PointLight.new_object(0x808080)
    light1.position.set(10., 10., 10.)
    light2 = context.THREE.AmbientLight.new_object(0x808080)
    lights = (light1, light2)

    material = context.THREE.MeshLambertMaterial.new_object({
        'color': 0xfffacd,
    })

    material_deformed = context.THREE.MeshLambertMaterial.new_object({
        'color': 0xf0fff0,
    })

    camera = threejs_visualization.build_perspective_camera(
        field_of_view=30, position=(10.0, 10.0, 10.0))

    mesh = {'vertices': vertices, 'faces': faces, 'material': material}
    transformed_mesh = {
        'vertices': my_vertices,
        'faces': faces,
        'material': material_deformed
    }
    geometries = threejs_visualization.triangular_mesh_renderer(
        [mesh, transformed_mesh],
        lights=lights,
        camera=camera,
        width=400,
        height=400)

    self.geometries = geometries

  def update(self, transformed_points):
    self.geometries[1].getAttribute('position').copyArray(
        transformed_points.numpy().ravel().tolist())
    self.geometries[1].getAttribute('position').needsUpdate = True
````

````
def get_random_transform():
  # Forms a random translation
  with tf.name_scope('translation_variable'):
    random_translation = tf.Variable(
        np.random.uniform(-2.0, 2.0, (3,)), dtype=tf.float32)

  # Forms a random quaternion
  hi = np.pi
  lo = -hi
  random_angles = np.random.uniform(lo, hi, (3,)).astype(np.float32)
  with tf.name_scope('rotation_variable'):
    random_quaternion = tf.Variable(quaternion.from_euler(random_angles))

  return random_quaternion, random_translation
````

````
random_quaternion, random_translation = get_random_transform()

initial_orientation = transform_points(vertices, random_quaternion,
                                       random_translation).numpy()
viewer = Viewer(initial_orientation)

predicted_transformation = model.predict(initial_orientation[tf.newaxis, :, :])

predicted_inverse_q = quaternion.inverse(predicted_transformation[0, 0:4])
predicted_inverse_t = -predicted_transformation[0, 4:]

predicted_aligned = quaternion.rotate(initial_orientation + predicted_inverse_t,
                                      predicted_inverse_q)

viewer = Viewer(predicted_aligned)
````

### Optimización matemática
Ahora resolveremos el problema desde un enfoque de optimización, que es un método tradicional para el problema de estimación de posición. Dada la correspondencia entre el objeto en posición resposo y su contraparte transformada, el problema se puede formular como un problema de minimización. La función de pérdida puede por ejemplo ser definida como la suma de las distancias euclideanas entre los puntos correspondientes.

Después, uno puede calcular la derivada de los parámetros de rotación y traslación respecto a la función de error, y seguir en la dirección del gradiente hasta convergir.

Cabe aclarar que aunque los resultados son buenos, hay métodos más eficientes para resolver este problema específico, como el algoritmo Kabsch.

Definimos la función de loss y de gradient.

````
def loss(target_points, quaternion_variable, translation_variable):
  transformed_points = transform_points(target_points, quaternion_variable,
                                        translation_variable)
  error = (vertices - transformed_points) / num_vertices
  return vector.dot(error, error)


def gradient_loss(target_points, quaternion, translation):
  with tf.GradientTape() as tape:
    loss_value = loss(target_points, quaternion, translation)
  return tape.gradient(loss_value, [quaternion, translation])
````

Optimizador

````
learning_rate = 0.05
with tf.name_scope('optimization'):
  optimizer = tf.compat.v1.train.AdamOptimizer(learning_rate)
````

Inicializamos transformación aleatoria y animamos el resultado

````
random_quaternion, random_translation = get_random_transform()

transformed_points = transform_points(vertices, random_quaternion,
                                      random_translation)

viewer = Viewer(transformed_points)

nb_iterations = 100
for it in range(nb_iterations):
  gradients_loss = gradient_loss(vertices, random_quaternion,
                                 random_translation)
  optimizer.apply_gradients(
      zip(gradients_loss, (random_quaternion, random_translation)))
  transformed_points = transform_points(vertices, random_quaternion,
                                        random_translation)

  viewer.update(transformed_points)
  time.sleep(0.1)
````

# Modelación de cámaras

Las cámaras son un hardware complejo que logran capturar imágenes 2D de objetos 3D. Para poder modelar cámaras precisamentes, se deben tomar en cuenta muchas cosas como la distorción de los lentes, la sensitividad del sensor de la cámara, la longitud focal y el tiempo de exposición. En este ejercicio nos enfocaremos en el modelo proyectivo de las cámaras.

![](https://storage.googleapis.com/tensorflow-graphics/notebooks/intrinsics/camera_model.jpg)

Este modelo está compuesto por dos parámetros normalmente referidos como parámetros intrínsecos:
* el punto principal, que es la proyección del centro óptico de la imagen. Idealmente, este punto debe estar cerca al centro de la imagen.
* la longitud focal, que es la distancia entre el centro óptico y el plano de la imagen. Estos parámetros permiten controlar el nivel del zoom.

Este ejercicio ilustra cómo utilizar TensorFlow Graphics para estimar estos parámetros intrínsicos. Poder estimar estos parámetros es importante para varias tareas, como reconstrucción 3D.

El objetivo será recuperar los parámetros intrínsecos dada una observación y las correspondencias entre la observación y el render. Las cosas se mantendrán sencillas: sólo insertaremos un rectángulo en una escena 3D.


````
import tensorflow_graphics.rendering.camera as tfg_camera
from tensorflow_graphics.math.optimizer import levenberg_marquardt
from tensorflow_graphics.rendering.camera import perspective  
````

## Ejemplo interactivo de este modelo 

````
def render_rectangle(rectangle_vertices, focal, principal_point, image_dimensions):
  """Renders un rectángulo en el plano de la imagen

  Args:
    rectangle_vertices: cuatro esquinas 3D del rectángulo.
    focal: longitud focal.
    principal_point: posición del punto principal en el plano de la imagen
    image_dimensions: dimensiones en pixeles de la imagen

  Returns:
    Una imagen 2D del rectángulo 3D.
  """
  image = np.zeros((int(image_dimensions[0]), int(image_dimensions[1]), 3))
  vertices_2d = perspective.project(rectangle_vertices, focal, principal_point)
  vertices_2d_np = vertices_2d.numpy()
  top_left_corner = np.maximum(vertices_2d_np[0, :], (0, 0)).astype(int)
  bottom_right_corner = np.minimum(
      vertices_2d_np[1, :],
      (image_dimensions[1] - 1, image_dimensions[0] - 1)).astype(int)
  for x in range(top_left_corner[0], bottom_right_corner[0] + 1):
    for y in range(top_left_corner[1], bottom_right_corner[1] + 1):
      c1 = float(bottom_right_corner[0] + 1 -
                 x) / float(bottom_right_corner[0] + 1 - top_left_corner[0])
      c2 = float(bottom_right_corner[1] + 1 -
                 y) / float(bottom_right_corner[1] + 1 - top_left_corner[1])
      image[y, x] = (c1, c2, 1)
  return image
````

````
# Vértices de un rectángulo
rectangle_depth = 1000.0
rectangle_vertices = np.array(
    ((-150.0, -75.0, rectangle_depth), (150.0, 75.0, rectangle_depth)))

# Tamaño del plano de la imagen
image_width = 400
image_height = 300
image_dimensions = np.array((image_height, image_width), dtype=np.float64)

# Usa la misma longitud focal horizontal y vertical. 
focal_lengths = np.array((image_height, image_height), dtype=np.float64)

# Pone el punto principal en el centro de la imagen.
ideal_principal_point = np.array(
    (image_width, image_height), dtype=np.float64) / 2.0

# Muestra la escena con los parámetros intrínsecos definidos arriba.
render = render_rectangle(rectangle_vertices, focal_lengths, ideal_principal_point,
                          image_dimensions)
_ = plt.imshow(render)
````

## Optimización de parámetros intrínsecos

Cualquier cámara viene con sus propios parámetros intrínsecos. Como se mencionó antes, tener los parámetros intrínsecos precisos sirven para reconstrucción de escenas 3D, robótica y sistemas de navegación.

Una manera común de estimar estos parámetros es utilizando un objeto 3D conocido. Sabiendo nuestra estimación actual de los parámetros intrínsecos, se puede predecir cómo se "verá" un objeto 3D y compararlo con la observación actual.

Iniciaremos definiendo algunas funciones de ayuda para visualizar los resultados.

````
def plot_optimization_step(observation, prediction):
  plt.figure(figsize=(20, 10))
  ax = plt.subplot("131")
  ax.set_title("Observation")
  _ = ax.imshow(observation)
  ax = plt.subplot("132")
  ax.set_title("Prediction using estimated intrinsics")
  _ = ax.imshow(prediction)
  ax = plt.subplot("133")
  ax.set_title("Difference image")
  _ = ax.imshow(np.abs(observation - prediction))
  plt.show()


def print_errors(focal_error, center_error):
  print("Error focal length %f" % (focal_error,))
  print("Err principal point %f" % (center_error,))
````

Let's now define the values of the intrinsic parameters we are looking for, and an initial guess of the values of these parameters.

Ahora definimos los valores de los parámetros intrínsecos que estamos buscando y la estimación inicial de los valores de estos parámetros.

````
def build_parameters():
  # Construye los parámetros intrínsecos que queremos recuperar.
  real_focal_lengths = focal_lengths * np.random.uniform(0.8, 1.2, size=(2,))
  real_principal_point = ideal_principal_point + (np.random.random(2) -
                                                  0.5) * image_width / 5.0

  # Inicializa la primera estimación de los parámetros intrínsecos.
  estimate_focal_lengths = tf.Variable(real_focal_lengths +
                                       (np.random.random(2) - 0.5) *
                                       image_width)
  estimate_principal_point = tf.Variable(real_principal_point +
                                         (np.random.random(2) - 0.5) *
                                         image_width / 4)
  return real_focal_lengths, real_principal_point, estimate_focal_lengths, estimate_principal_point
````

Como se describió antes, uno puede comparar cómo se verían los objetos 3D usando la estimación actual y la observación actual. La siguiente función captura la distancia entre las dos imágenes. El objetivo será minimizar esta función.

````
def residuals(estimate_focal_lengths, estimate_principal_point):
  vertices_2d_gt = perspective.project(rectangle_vertices, real_focal_lengths,
                                       real_principal_point)
  vertices_2d_observed = perspective.project(rectangle_vertices,
                                             estimate_focal_lengths,
                                             estimate_principal_point)
  return vertices_2d_gt - vertices_2d_observed
````

Todas las piezas están en lugar para resolver el problema. 

Nota: los residuales se minimizan utilizando el algoritmo Levenberg-Marquardt, el cuál es especialmente apropiado para este problema. También se podrían utilizar optimizadores de primer órden (Adam o Gradient Descent).

````
# Crea samples de parámetros intrínsecos a recuperar y la solución inicial.
real_focal_lengths, real_principal_point, estimate_focal_lengths, estimate_principal_point = build_parameters(
)

# Construye imagen observada.
observation = render_rectangle(rectangle_vertices, real_focal_lengths,
                               real_principal_point, image_dimensions)

# Muestra solución inicial
print("Configuración inicial:")
print_errors(
    np.linalg.norm(estimate_focal_lengths - real_focal_lengths),
    np.linalg.norm(estimate_principal_point - real_principal_point))
image = render_rectangle(rectangle_vertices, estimate_focal_lengths,
                         estimate_principal_point, image_dimensions)
plot_optimization_step(observation, image)

# Optimización inicial.
_, (estimate_focal_lengths,
    estimate_principal_point) = levenberg_marquardt.minimize(
        residuals, (estimate_focal_lengths, estimate_principal_point), 1)

print("Predicted configuration:")
print_errors(
    np.linalg.norm(estimate_focal_lengths - real_focal_lengths),
    np.linalg.norm(estimate_principal_point - real_principal_point))
image = render_rectangle(rectangle_vertices, estimate_focal_lengths,
                         estimate_principal_point, image_dimensions)
plot_optimization_step(observation, image)
````

# Materiales y luces

El mundo alrededor de nosotros es muy complejo y está compuesto de muchos materiales desde cristal a madera. Cada material tiene sus propias propiedades intrínsecas e interactúan de manera diferente con la luz. Por ejemplo, algunos materiales son difusos (papel o marmol) y dadas las condiciones adecuadas de luz, se ven igual desde cualquier ángulos. Otros materiales como el metal tienen una apariencia que cambia significativamente y exhibe efectos como especuliaridades.

Modelar exactamente cómo la luz interactúa con los materiales es un proceso complejo que involucra diferentes effectos. En este ejercicio, nos enfocaremos en el efecto más común que es la reflectancia. [Bidirectional reflectance distribution functions (BRDF)](https://en.wikipedia.org/wiki/Bidirectional_reflectance_distribution_function) o [Función de distribución de reflectancia bidireccional](https://es.wikipedia.org/wiki/Funci%C3%B3n_de_distribuci%C3%B3n_de_reflectancia_bidireccional) es el método más común de modelar reflectancia. Dada la dirección de la luz entrante, BRDFs controlan la cantidad de luz que rebota en la dirección en la cuál la superficie está siendo observada (vectores grises abajo).

![](https://storage.googleapis.com/tensorflow-graphics/notebooks/reflectance/brdf.jpg)

En este ejercicio, la luz se lanza a tres esferas, cada una con un tipo de material descrito en la imagen de arriba. El material especular se modela con el [Phong specular model]

In this Colab, a light we be shone onto three spheres, each with a material described in the image above, where the specular material is going to be modelled with the [Phong specular model](https://en.wikipedia.org/wiki/Phong_reflection_model).

**Nota**: Este Colab cubre un tema avanzada y, por lo tanto, se enfoca en proveer un ejemplo controlable para formar un entendimiento de alto nivel de BRDFs en vez de proveer explicaciones paso a paso.

````
import math as m
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf

from tensorflow_graphics.rendering.reflectance import lambertian
from tensorflow_graphics.rendering.reflectance import phong
from tensorflow_graphics.rendering.camera import orthographic
from tensorflow_graphics.geometry.representation import grid
from tensorflow_graphics.geometry.representation import ray
from tensorflow_graphics.geometry.representation import vector
````

````
###############
# UI controls #
###############
#@title Controls { vertical-output: false, run: "auto" }
light_x_position = -0.4  #@param { type: "slider", min: -1, max: 1 , step: 0.05 }
albedo_red = 0.1  #@param { type: "slider", min: 0.0, max: 1.0 , step: 0.1 }
albedo_green = 1  #@param { type: "slider", min: 0.0, max: 1.0 , step: 0.1 }
albedo_blue = 0.5  #@param { type: "slider", min: 0.0, max: 1.0 , step: 0.1 }
light_red = 1  #@param { type: "slider", min: 0.0, max: 1.0 , step: 0.1 }
light_green = 0.3  #@param { type: "slider", min: 0.0, max: 1.0 , step: 0.1 }
light_blue = 0.7  #@param { type: "slider", min: 0.0, max: 1.0 , step: 0.1 }
specular_percentage = 0.57  #@param { type: "slider", min: 0, max: 1 , step: 0.01 }
shininess = 8  #@param { type: "slider", min: 0, max: 10, step: 1 }
diffuse_percentage = 1.0 - specular_percentage
dtype = np.float64
albedo = np.array((albedo_red, albedo_green, albedo_blue), dtype=dtype)

def compute_intersection_normal_sphere(image_width, image_height, sphere_radius,
                                       sphere_center, dtype):
  pixel_grid_start = np.array((0.5, 0.5), dtype=dtype)
  pixel_grid_end = np.array((image_width - 0.5, image_height - 0.5), dtype=dtype)
  pixel_nb = np.array((image_width, image_height))
  pixels = grid.generate(pixel_grid_start, pixel_grid_end, pixel_nb)

  pixel_ray = tf.math.l2_normalize(orthographic.ray(pixels), axis=-1)
  zero_depth = np.zeros([image_width, image_height, 1])
  pixels_3d = orthographic.unproject(pixels, zero_depth)

  intersections_points, normals = ray.intersection_ray_sphere(
      sphere_center, sphere_radius, pixel_ray, pixels_3d)
  intersections_points = np.nan_to_num(intersections_points)
  normals = np.nan_to_num(normals)
  return intersections_points[0, :, :, :], normals[0, :, :, :]

#####################################
# Setup the image, sphere and light #
#####################################
# Image dimensions
image_width = 400
image_height = 300

# Sphere center and radius
sphere_radius = np.array((100.0,), dtype=dtype)
sphere_center = np.array((image_width / 2.0, image_height / 2.0, 300.0),
                         dtype=dtype)

# Set the light along the image plane
light_position = np.array((image_width / 2.0 + light_x_position * image_width,
                           image_height / 2.0, 0.0),
                          dtype=dtype)
vector_light_to_sphere_center = light_position - sphere_center
light_intensity_scale = vector.dot(
    vector_light_to_sphere_center, vector_light_to_sphere_center,
    axis=-1) * 4.0 * m.pi
light_intensity = np.array(
    (light_red, light_green, light_blue)) * light_intensity_scale

################################################################################################
# For each pixel in the image, estimate the corresponding surface point and associated normal. #
################################################################################################
intersection_3d, surface_normal = compute_intersection_normal_sphere(
    image_width, image_height, sphere_radius, sphere_center, dtype)

#######################################
# Reflectance and radiance estimation #
#######################################
incoming_light_direction = tf.math.l2_normalize(
    intersection_3d - light_position, axis=-1)
outgoing_ray = np.array((0.0, 0.0, -1.0), dtype=dtype)
albedo = tf.broadcast_to(albedo, tf.shape(surface_normal))

# Lambertian BRDF
brdf_lambertian = diffuse_percentage * lambertian.brdf(incoming_light_direction, outgoing_ray,
                                  surface_normal, albedo)
# Phong BRDF
brdf_phong = specular_percentage * phong.brdf(incoming_light_direction, outgoing_ray, surface_normal,
                        np.array((shininess,), dtype=dtype), albedo)
# Composite BRDF
brdf_composite = brdf_lambertian + brdf_phong
# Irradiance
cosine_term = vector.dot(surface_normal, -incoming_light_direction)
cosine_term = tf.math.maximum(tf.zeros_like(cosine_term), cosine_term)
vector_light_to_surface = intersection_3d - light_position
light_to_surface_distance_squared = vector.dot(
    vector_light_to_surface, vector_light_to_surface, axis=-1)
irradiance = light_intensity / (4 * m.pi *
                                light_to_surface_distance_squared) * cosine_term
# Rendering equation
zeros = tf.zeros(intersection_3d.shape)
radiance = brdf_composite * irradiance
radiance_lambertian = brdf_lambertian * irradiance
radiance_phong = brdf_phong * irradiance

###############################
# Display the rendered sphere #
###############################
# Saturates radiances at 1 for rendering purposes.
radiance = np.minimum(radiance, 1.0)
radiance_lambertian = np.minimum(radiance_lambertian, 1.0)
radiance_phong = np.minimum(radiance_phong, 1.0)
# Gamma correction
radiance = np.power(radiance, 1.0 / 2.2)
radiance_lambertian = np.power(radiance_lambertian, 1.0 / 2.2)
radiance_phong = np.power(radiance_phong, 1.0 / 2.2)

plt.figure(figsize=(20, 20))

# Diffuse
radiance_lambertian = np.transpose(radiance_lambertian, (1, 0, 2))
ax = plt.subplot("131")
ax.axes.get_xaxis().set_visible(False)
ax.axes.get_yaxis().set_visible(False)
ax.grid(False)
ax.set_title("Lambertian")
_ = ax.imshow(radiance_lambertian)

# Specular
radiance_phong = np.transpose(radiance_phong, (1, 0, 2))
ax = plt.subplot("132")
ax.axes.get_xaxis().set_visible(False)
ax.axes.get_yaxis().set_visible(False)
ax.grid(False)
ax.set_title("Specular - Phong")
_ = ax.imshow(radiance_phong)

# Diffuse + specular
radiance = np.transpose(radiance, (1, 0, 2))
ax = plt.subplot("133")
ax.axes.get_xaxis().set_visible(False)
ax.axes.get_yaxis().set_visible(False)
ax.grid(False)
ax.set_title("Combined lambertian and specular")
_ = ax.imshow(radiance)
````

## Convoluciones de imagen vs convoluciones de grafo

Las imágenes están representadas por grids (cuadrículas) uniformes de pixeles. Ejecutar convoluciones en grids es un proceso bien entendido y central en muchos productos e investigaciones académicas

![](https://storage.googleapis.com/tensorflow-graphics/notebooks/mesh_segmentation/cat_image_convolutions.png)

Sin embargo, las cosas se vuelven más complicadas cuando se trata de objetos tridimensionales como meshes o point clouds dado que estos no están definidos en grids regulares. Una operación convolucional para meshes o point clouds debe operar en estructuras de datos irregulares, lo cuál hace que las CNNs basadas en estas sean más difíciles de implementar.

![](https://storage.googleapis.com/tensorflow-graphics/notebooks/mesh_segmentation/cat_mesh_convolutions.png)

En general, un mesh se puede representar como un grafo que no está limitado a un grid regular. Varios operadores convolucionales de grafo se han publicado en los años recientes. En este ejercicio, utilizaremos [Feature Steered Graph Convolutions](https://arxiv.org/abs/1706.05206). Similar a su contraparte de imágenes, este bloque básico de construcción sirve para muchas tareas.

### Cargar y visualizar datos de prueba

Para convoluciones de grafos, necesitamos un *matriz de adjacencia con pesos* (weighted adjacency matrix) que denote la conectividad del mes. Las convoluciones Feature-steered esperan que los vértices estén contectados a sí mismos (arista a sí mismos). Es decir, la matriz de la matriz de adjacencia no puede ser cero. La matriz se define:

```
A[i, j] = w[i,j] si el vértice i y el vértice j comparten una arista.
A[i, i] = w[i,i] para cada vértices i.
A[i, j] = 0 sino.
donde, w[i, j] = 1/(degree(vertex i)), y sum(j)(w[i,j]) = 1
```

Aquí, degree(vertex i) es el número de aristas de un vértice (inclyendo la arista a sí mismo). La matriz de adyacencia se guarda como un SparseTensor.

Cargamos los meshes de prueba usando el [tf.data.TFRecordDataset](https://www.tensorflow.org/api_docs/python/tf/data/TFRecordDataset) descargado arriba. Cada mesh es guardado como un [tf.Example](https://www.tensorflow.org/api_docs/python/tf/train/Example) con los siguientes campos:

*   'num_vertices': Número de vértices en el mesh. 
*   'num_triangles': Número de triángulos en el mesh. 
*   'vertices': A [V, 3] float tensor con las posiciones de los vértices.
*   'triangles': A [T, 3] integer tensor con los índices de los vértices de cada triángulo.
*   'labels': A [V] integer tensor con la etiqueta de la clase de segmentación para cada vértice.
Donde 'V' es el número de vértices y 'T' es el número de triángulos para cada mesh. Como cada mesh puede tener un número diferente de vértices y caras, hacemos un pad a los tensores de datos con '0's en cada batch.

### Definición del modelo

Dado un mesh con sus vértices V y los features de entrada de D dimensiones (posición de vértice, normal, etc), nos gustaría crear un modelo capaz de clasificar si el vértice corresponde a una clase. Primero crearemos un codificador (encoder) que codifica cada vértice en el mesh a logits de C dimensiones, donde C es el número de partes. Primero utilizaremos convoluciones 1x1 para cambiar las dimensiones de los features de entrada. Luego aplicamos una secuencia de convoluciones de grafo y no linearidades ReLU. Finalmente, aplicamos convoluciones 1x1, las cuales se utilizan para computar el softmax cross entropy descrito abajo. 

Nota que este modelo no utiliza ninguna técnica de pooling, que está fuera del alcance del ejercicio.

![](https://storage.googleapis.com/tensorflow-graphics/notebooks/mesh_segmentation/mesh_segmentation_model_def.png)

````
MODEL_PARAMS = {
    'num_filters': 8,
    'num_classes': 16,
    'encoder_filter_dims': [32, 64, 128],
}


def mesh_encoder(batch_mesh_data, num_filters, output_dim, conv_layer_dims):
  """Un encoder de mesh que utiliza convoluciones de grafo feature steered.

    Las abreviaciones usadas abajo son:
      `B`: Tamaño de batch.
      `V`: Máximo número de vértices de todos los vértices en un batch.
      `D`: El número de dimensiones de los features de los vértices de entrada.
           D=3 si sólo se utilizan las posiciones de los vértices como features.

  Args:
    batch_mesh_data: Un diccionario mesh_data con las siguientes llaves
      'vertices': A [B, V, D] `float32` tensor con los features de cada vértice.
      'neighbors': A [B, V, V] `float32` sparse tensor con pesos de arista.
      'num_vertices': A [B] `int32` tensor con número de vértices en el mesh.
    num_filters: Número de matrices de peso a ser usadas en la convolución.
    output_dim: La dimensión de salida de los features de salida.
    conv_layer_dims: Lista de las dimensiones de las capas convolucionales.

  Returns:
    vertex_features: A [B, V, output_dim] `float32` tensor con los features de cada vértice
  """
  batch_vertices = batch_mesh_data['vertices']

  # Linear: N x D --> N x 16.
  vertex_features = tf.keras.layers.Conv1D(16, 1, name='lin16')(batch_vertices)

  # capas convolucionares
  for dim in conv_layer_dims:
    with tf.variable_scope('conv_%d' % dim):
      vertex_features = graph_conv.feature_steered_convolution_layer(
          vertex_features,
          batch_mesh_data['neighbors'],
          batch_mesh_data['num_vertices'],
          num_weight_matrices=num_filters,
          num_output_channels=dim)
    vertex_features = tf.nn.relu(vertex_features)

  # Linear: N x 128 --> N x 256.
  vertex_features = tf.keras.layers.Conv1D(
      256, 1, name='lin256')(
          vertex_features)
  vertex_features = tf.nn.relu(vertex_features)

  # Linear: N x 256 --> N x output_dim.
  vertex_features = tf.keras.layers.Conv1D(
      output_dim, 1, name='lin_output')(
          vertex_features)

  return vertex_features
````

Dado un codificador de mesh, vamos a definir un `model_fn`para un [tf.Estimator](https://www.tensorflow.org/api_docs/python/tf/estimator/Estimator) para la clasificación de vértices utilizando la pérdida softmax cross entropy. Un tf.Estimator model_fn regresa las operacionesn ecesarias para realizar entrenamiento, evaluación o predicciones dadas las entradas y otros parámetros.

````
def get_learning_rate(params):
  """Regresa el decaying learning rate."""
  global_step = tf.train.get_or_create_global_step()
  learning_rate = tf.train.exponential_decay(
      params['init_learning_rate'],
      global_step,
      params['lr_decay_steps'],
      params['lr_decay_rate'])
  return learning_rate

def model_fn(features, labels, mode, params):
  """Regresa el model_fn de segmentación de mesh a ser usado con tf.Estimator."""
  logits = mesh_encoder(features, params['num_filters'], params['num_classes'],
                        params['encoder_filter_dims'])
  predictions = tf.argmax(logits, axis=-1, output_type=tf.int32)
  outputs = {
      'vertices': features['vertices'],
      'triangles': features['triangles'],
      'num_vertices': features['num_vertices'],
      'num_triangles': features['num_triangles'],
      'predictions': predictions,
  }
  # Para prediccions, regresa la salida.
  if mode == tf.estimator.ModeKeys.PREDICT:
    outputs['labels'] = features['labels']
    return tf.estimator.EstimatorSpec(mode=mode, predictions=outputs)
  # Pérdida
  vertex_ragged_sizes = features['num_vertices']
  mask = tf.sequence_mask(vertex_ragged_sizes, tf.shape(labels)[-1])
  loss_weights = tf.cast(mask, dtype=tf.float32)
  loss = tf.losses.sparse_softmax_cross_entropy(
      logits=logits, labels=labels, weights=loss_weights)
  # Construye el optimizador para entrenamiento
  if mode == tf.estimator.ModeKeys.TRAIN:
    optimizer = tf.train.AdamOptimizer(
        learning_rate=get_learning_rate(params),
        beta1=params['beta'],
        epsilon=params['adam_epsilon'])
    update_ops = tf.get_collection(tf.GraphKeys.UPDATE_OPS)
    with tf.control_dependencies(update_ops):
      train_op = optimizer.minimize(
          loss=loss, global_step=tf.train.get_global_step())
    return tf.estimator.EstimatorSpec(mode=mode, loss=loss, train_op=train_op)

  # Para evaluación regresa las métricas.
  eval_ops = {
      'mean_loss':
          tf.metrics.mean(loss),
      'accuracy':
          tf.metrics.accuracy(
              labels=labels, predictions=predictions, weights=loss_weights)
  }
  return tf.estimator.EstimatorSpec(
      mode=mode, loss=loss, eval_metric_ops=eval_ops)
````

### Entrenar el modelo desde cero.

Ahora vamos a entrenar el modelo desde 0 usando el estimador que definimos más arriba.

````
path_to_train_data_zip = tf.keras.utils.get_file(
    'train_data.zip',
    origin='https://storage.googleapis.com/tensorflow-graphics/notebooks/mesh_segmentation/train_data.zip',
    extract=True)

train_data_files = glob.glob(
    os.path.join(os.path.dirname(path_to_train_data_zip), '*train*.tfrecords'))

retrain_model_dir = os.path.join(local_model_dir, 'retrain')
````

````
train_io_params = {
    'batch_size': 8,
    'parallel_threads': 8,
    'is_training': True,
    'shuffle': True,
    'sloppy': True,
}

eval_io_params = {
    'batch_size': 8,
    'parallel_threads': 8,
    'is_training': False,
    'shuffle': False
}


def train_fn():
  return dataio.create_input_from_dataset(dataio.create_dataset_from_tfrecords,
                                          train_data_files, train_io_params)


def eval_fn():
  return dataio.create_input_from_dataset(dataio.create_dataset_from_tfrecords,
                                          test_data_files, eval_io_params)


train_params = {
    'beta': 0.9,
    'adam_epsilon': 1e-8,
    'init_learning_rate': 0.001,
    'lr_decay_steps': 10000,
    'lr_decay_rate': 0.95,
}

train_params.update(MODEL_PARAMS)

checkpoint_delay = 120  # Checkpoint every 2 minutes.
max_steps = 100000  # Number of training steps.

config = tf.estimator.RunConfig(
    log_step_count_steps=1,
    save_checkpoints_secs=checkpoint_delay,
    keep_checkpoint_max=3)

classifier = tf.estimator.Estimator(
    model_fn=model_fn,
    model_dir=retrain_model_dir,
    config=config,
    params=train_params)
train_spec = tf.estimator.TrainSpec(input_fn=train_fn, max_steps=max_steps)
eval_spec = tf.estimator.EvalSpec(
    input_fn=eval_fn,
    steps=None,
    start_delay_secs=2 * checkpoint_delay,
    throttle_secs=checkpoint_delay)

print('Start training & eval.')
tf.estimator.train_and_evaluate(classifier, train_spec, eval_spec)
print('Train and eval done.')
````
